# Phase 6: Model testing and evaluation
## Overview
In this final section I will be testing my model on my test data. I will also evaluate it on its r^2 value, and a visual inspection. I will also try picking a coordinate in the USA myself and making sure the model works.

## Import needed libraries and files

```python
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error
import numpy as np
from sklearn.metrics import r2_score
import joblib
import matplotlib.pyplot as plt

training_df = pd.read_csv('../processed_data/dataset_splits/training_set.csv')
validation_df = pd.read_csv('../processed_data/dataset_splits/validation_set.csv')
model = joblib.load('../models/old_model.joblib')
```

## Running the model on the test data
Now to run the model on the test data and find the r^2 value

```python
test_df = pd.read_csv('../processed_data/dataset_splits/test_set.csv')
test_input = test_df.drop(columns=['performance_ratio'])
test_output = test_df['performance_ratio']

test_output_prediction = model.predict(test_input)
final_test_rmse = np.sqrt(mean_squared_error(test_output, test_output_prediction))

print("The RMSE of the test: " + str(final_test_rmse))

# Calculate the R-squared score
r2 = r2_score(test_output, test_output_prediction)
print("The r^2 of the test: " + str(r2))
```

## Visual analysis
Next I'm going to plot the data to see how my model is performed. First I'm going to compare predicted vs actual outcome.

```python
x_values = []
y_values = []
for i in range(178):
    x_values.append(i+1)
    y_values.append(abs(test_output_prediction[i] - test_output[i]))


plt.figure(figsize=(14, 5))
plt.scatter(x_values, y_values, color='blue', marker='o')

plt.ylabel('Absolute estimation error')
plt.xlabel('Index')
plt.grid(True, linestyle=':', alpha=0.6)
plt.title('Absolute difference between predicted and actual performance ratio')
plt.show()
```

```python
y_values = []
x_values = []
for i in range(178):
    x_values.append(i+1)
    y_values.append(abs(test_output[i]))

plt.figure(figsize=(14, 5))
plt.scatter(x_values, y_values, color='red', marker='o')

plt.ylabel('Performance ratio')
plt.xlabel('Index')
plt.grid(True, linestyle=':', alpha=0.6)
plt.title('Absolute difference between predicted and actual performance ratio')
plt.show()
```

It's hard to see but it looks like the worst predictions of my model tended to have outlier actual outputs. To test this hypothesis I'm going to compare the absolute distance from the mean performance ratio to the absolute difference between predicted and actual performance. I'm going to also find the covariance between the two variables

```python
x_values = []
y_values = []
for i in range(178):
    x_values.append(abs(test_output[i] - test_output.mean()))
    y_values.append(abs(test_output_prediction[i] - test_output[i]))

print("The covariance is: " + str(np.corrcoef(x_values, y_values)[0, 1]))

plt.figure(figsize=(14, 5))
plt.scatter(x_values, y_values, color='blue', marker='o')

plt.title('Absolute distance from the mean performance ratio vs absolute difference between predicted and actual performance')
plt.xlabel('Distance from the mean performance ratio')
plt.ylabel('Absolute difference between predicted and actual performance')
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()
```

## Conclusion
The model got a r^2 of 0.52 which is a fair bit better than just picking the average value but far from perfect. Looking at where my model went wrong it looks like at extreme values it didn't perform aswell, this is shown by the strong correlation between extreme true performance ratios and weak predictions. This is likely to be because my model hasn't seen enough of these extreme examples so more data would be needed to create a better model.
