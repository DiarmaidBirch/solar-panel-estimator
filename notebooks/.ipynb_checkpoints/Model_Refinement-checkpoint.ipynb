```python
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score
import numpy as np

training_df = pd.read_csv('../processed_data/dataset_splits/training_set.csv')
validation_df = pd.read_csv('../processed_data/dataset_splits/validation_set.csv')
display(training_df.head())
```

```python
# Sets the columns for the features and label for training
training_input = training_df.drop(columns=['performance_ratio', 'longitude'])
training_output = training_df['performance_ratio']

# Sets the columns for the features and label for validation
validation_input = validation_df.drop(columns=['performance_ratio', 'longitude'])
validation_output = validation_df['performance_ratio']

# The pipeline that will run the model
ridge_model = Pipeline([
    ('scaler', StandardScaler()),
    ('poly', PolynomialFeatures()),
    ('ridge', Ridge(random_state=42)) # Use Ridge instead of LinearRegression
])


# Test a range of alpha values (must be positive)
# np.logspace creates exponentially spaced values, which is good for alpha
#alphas_to_test = np.logspace(-4, 0, 5) # Example: [0.0001, 0.001, 0.01, 0.1, 1.0]
alphas_to_test = [0.1, 1.0, 10]

best_score = float('inf')
best_degree = 0
best_alpha = 0
best_r2 = 0

print("Tuning Ridge Regression: Degree x Alpha")
# --- 3. Nested Tuning Loop ---
for i in range(1, 6):
    for alpha in alphas_to_test:
        
        # Set both parameters in the pipeline
        ridge_model.set_params(poly__degree=i, ridge__alpha=alpha)
        
        # Train and Predict
        ridge_model.fit(training_input, training_output)
        predicted_validation_output = ridge_model.predict(validation_input)
        
        # Evaluate performance (RMSE)
        rmse = np.sqrt(mean_squared_error(validation_output, predicted_validation_output))
        
        #print(f"  Degree {degree}, Alpha {alpha:.4f}: RMSE = {rmse:.4f}")

        #print(str(i) + " " + str(alpha) + " " + str(rmse))

        r2 = r2_score(validation_output, predicted_validation_output)
        
        # Update the best model
        #if rmse < best_score:
        if r2 > best_r2:
            best_r2 = r2
            best_score = rmse
            best_degree = i
            best_alpha = alpha

print("The best model has degree: " + str(best_degree) + ", alpha: " + str(best_alpha) + ", with RMSE: " + str(best_r2))
```

```python
# Sets the columns for the features and label for training
combined_input = pd.concat([training_input, validation_input], axis=0, ignore_index=True)
combined_output = pd.concat([training_output, validation_output], axis=0, ignore_index=True)

ridge_model = Pipeline([
    ('scaler', StandardScaler()),
    ('poly', PolynomialFeatures()),
    ('ridge', Ridge(random_state=42)) # Use Ridge instead of LinearRegression
])

# Set both parameters in the pipeline
ridge_model.set_params(poly__degree=1, ridge__alpha=1)

# Train and Predict
ridge_model.fit(combined_input, combined_output)

####

test_df = pd.read_csv('../processed_data/dataset_splits/test_set.csv')
#display(test_df.head(50))

test_input = test_df.drop(columns=['performance_ratio', 'longitude'])
test_output = test_df['performance_ratio']

predicted_test_output = ridge_model.predict(test_input)
final_test_rmse = np.sqrt(mean_squared_error(test_output, predicted_test_output))

print("The RMSE of the test: " + str(final_test_rmse))


r2 = r2_score(test_output, predicted_test_output)
print("The r^2 of the test: " + str(r2))
```
